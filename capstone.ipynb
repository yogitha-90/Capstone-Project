```python
# # import pandas as pd

# # # Assuming df is your DataFrame and 'your_column' is the column with mixed data types
# # df['your_column'] = pd.to_numeric(df['your_column'], errors='coerce')

# # # Drop rows with NaN values in 'your_column'
# # df.dropna(subset=['your_column'], inplace=True)
df_train['sourceId'] = pd.to_numeric(df_train['sourceId'], errors='coerce')
for sourceId in df_train.columns:
    unique_values = df_train['sourceId'].unique()
    # print(f"Unique values in {'sourceId'} : {unique_values}")
 # data_type = df['your_column'].dtype
 # print(f"The data type of 'your_column' is: {data_type}")
data_type = df_train['sourceId'].dtype
print(f"The data type of 'sourceId' is: {data_type}")
```

```python
df_train['locationId'] = pd.to_numeric(df_train['locationId'], errors='coerce')
df_train.dropna(subset=['locationId'], inplace=True)
for locationId in df_train.columns:
     unique_values = df_train['locationId'].unique()


data_type = df_train['locationId'].dtype
print(f"The data type of 'locationId' is: {data_type}")
```

```python
df_train['typeId'] = pd.to_numeric(df_train['typeId'], errors='coerce')
df_train.dropna(subset=['typeId'], inplace=True)
for locationId in df_train.columns:
     unique_values = df_train['typeId'].unique()
     

data_type = df_train['typeId'].dtype
print(f"The data type of 'typeId' is: {data_type}")
```

```python
df_train['price'] = pd.to_numeric(df_train['price'], errors='coerce')
df_train.dropna(subset=['price'], inplace=True)
for locationId in df_train.columns:
     unique_values = df_train['price'].unique()

data_type = df_train['price'].dtype
print(f"The data type of 'price' is: {data_type}")
```

```python


# # Assuming df is your DataFrame
# columns_to_convert = ['locationId', 'otherColumn1', 'otherColumn2']

# # Convert specified columns to numeric
# df[columns_to_convert] = df[columns_to_convert].astype(float)
columns_to_convert = ['rooms','bedrooms','bathrooms']
df_train[columns_to_convert] = df_train[columns_to_convert].astype(float)
new_column_names = ['rooms','bedrooms','bathrooms']
df_train = df_train.rename(columns=dict(zip(columns_to_convert, new_column_names)))
```
